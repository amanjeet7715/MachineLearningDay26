```python
import seaborn 
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

seaborn.set(style = 'whitegrid') 
fmri = seaborn.load_dataset("fmri") 
```

**x, y, hue : names of variables in data or vector data, optional
Inputs for plotting long-form data. See examples for interpretation.
data : DataFrame, array, or list of arrays, optional
Dataset for plotting. If x and y are absent, this is interpreted as wide-form. Otherwise it is expected to be long-form.**

```python
seaborn.boxplot(x ="timepoint", y ="signal", hue ="region", data = fmri) 
```

**orient : “v” | “h”, optional
Orientation of the plot (vertical or horizontal). This is usually inferred based on the type of the input variables, but it can be used to resolve ambiguitiy when both x and y are numeric or when plotting wide-form data.**

```python
seaborn.boxplot(x ="timepoint", y ="signal", hue ="region", data = fmri, orient='h') 
```

**color : matplotlib color, optional
Color for all of the elements, or seed for a gradient palette.**

```python
seaborn.boxplot(x ="timepoint", y ="signal", hue ="region", data = fmri, color='red') 
```

**palette : palette name, list, or dict
Colors to use for the different levels of the hue variable. Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors.**

```python
seaborn.boxplot(x ="timepoint", y ="signal", hue ="region", data=fmri, palette="Blues_d")
```

**dodge : bool, optional
When hue nesting is used, whether elements should be shifted along the categorical axis.**

```python
seaborn.boxplot(x ="timepoint", y ="signal", hue ="region", data = fmri, dodge=False) 
```
